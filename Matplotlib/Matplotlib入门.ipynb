```python
import matplotlib.pyplot as plt
import numpy as np
```

创建一个图像的最简单方法是使用pylot：

```python
fig = plt.figure()
fig.suptitle("No axes on this figure")

fig,ax_lst = plt.subplots(2,2)
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x,label='linar')
plt.plot(x,x**2,label='quadratic')
plt.plot(x,x**3,label='cubic')

plt.xlabel('X label')
plt.ylabel('Y label')

plt.title('Simple plot')

plt.legend()

plt.show()
```

```python
x = np.arange(0,10,0.2)
y = np.sin(x)
fig,ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
plt.plot([1,2,3,4,5,6,7,8,9])
plt.ylabel('same numbers')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'r:o')
plt.axis([0,6,0,20])
plt.show()
```

上例中的 axis() 命令采用 [xmin, xmax, ymin, ymax] 列表并指定轴的视口。

```python
t = np.arange(0.,5.,0.2)

plt.plot(t,t,'r--',t,t**2,'b-.',t,t**3,'g^')
plt.show()
```

```python
data = {
    'a':np.arange(50),
    'c':np.random.randint(0,50,50),
    'd':np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a','b',c='c',s='d',data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0,5.0,0.01)
s = np.cos(2*np.pi*t)
line,=plt.plot(t,s,lw=2)

plt.annotate('local max',xy=(2,1),xytext=(3,1.5),
            arrowprops=dict(facecolor='black',shrink=0.05),)

plt.ylim(-2,2)
plt.show()
```
