```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
#读取图像的原始数据
image_raw_data = tf.gfile.FastGFile("data/timg2.jpg",'rb').read()


```

```python
with tf.Session() as sess:
    # 对图像进行jpeg 的格式解码从而得到图像对应的三维矩阵。TensorFlow 还提供了
    # tf.image.decode_png 函数对png 格式的图像进行解码。解码之后的结果为一个
    # 张盘， 在使用它的取值之前需要明确调用运行的过程。
    img_data = tf.image.decode_jpeg(image_raw_data)
    
    #输出解码之后的三维矩阵，
    print(img_data.eval())
    
    #使用pyplot工具可视化得到的图像
    plt.imshow(img_data.eval())
    plt.show()
    
    #将表示一张图像的三维矩阵重新按照jpeg格式编码并存入文件中，打开这张图像
    encoded_image = tf.image.encode_jpeg(img_data)
    with tf.gfile.GFile("data/output_timg.jpg","wb") as f:
        f.write(encoded_image.eval())
```
