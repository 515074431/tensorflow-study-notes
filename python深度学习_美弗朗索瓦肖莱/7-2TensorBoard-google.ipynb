```python
import keras
from keras import layers
from keras.datasets import imdb
from keras.preprocessing import sequence
```

```python
max_features = 500 #原来2000 #作为特征的单词个数
max_len = 500

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_train = sequence.pad_sequences(x_train, maxlen=max_len)
x_test = sequence.pad_sequences(x_test, maxlen=max_len)

tensor_board_log = '/content/drive/My Drive/jupyter/tensorflow-study-notes/TensorBoardLog'
#tensor_board_log = '../TensorBoardLog'
model = keras.models.Sequential()
model.add(layers.Embedding(max_features, 128,
                          input_length=max_len,
                          name='embed'))
model.add(layers.Conv1D(32, 7, activation='relu'))
model.add(layers.MaxPooling1D(5))
model.add(layers.Conv1D(32, 7, activation='relu'))
model.add(layers.GlobalMaxPooling1D())
model.add(layers.Dense(1))
model.summary()
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['acc'])
```

我们用一个TensorBoard 回调函数实例来启动训练。这个回调函数会将日志事件写入硬
盘的指定位置


```python
callbacks = [
    keras.callbacks.TensorBoard(log_dir=tensor_board_log,#日志文件将被写入这个位置
                               histogram_freq=1,#每一轮之后记录激活直方图
                               embeddings_freq=1,#每一轮之后记录嵌入数据
                                embeddings_data =x_train[:128].astype('float32')
                               )
]

history = model.fit(x_train, y_train,
                   epochs=20,
                   batch_size=128,
                   validation_split=0.2,
                   callbacks=callbacks
                   )
```
