##2.7.3　示例：选择随机点





```
mean = [0,0]
cov = [[1,2],[2,5]]
X = rand.multivariate_normal(mean,cov,100)
X.shape
```

```
plt.scatter(X[:,0],X[:,1])
```

```
indices = np.random.choice(X.shape[0],20,replace=False)
indices
```

```
selection = X[indices]
selection.shape
```

现在来看哪些点被选中了，将选中的点在图上用大圆圈标示出来

```
plt.scatter(X[:,0],X[:,1],alpha=0.3)
plt.scatter(selection[:,0],selection[:,1],facecolor='none',edgecolor='b',s=200)
```

##2.7.5　示例：数据区间划分

```
np.random.seed(42)
x = np.random.randn(100)

#手动计算直方图
bins = np.linspace(-5,5,20)
counts = np.zeros_like(bins)

#为每个x找到合适的区间
i = np.searchsorted(bins,x)

#为每个区间加上1
np.add.at(counts,i,1)

#画出结果
plt.plot(bins,counts,linestyle='steps')
```

```
plt.hist(x,bins,histtype='step')
```

##2.8.1　NumPy中的快速排序：np.sort和np.argsort

如果想在不修改原始输入数组的基础上返回一个排好序的数组，可以使
用 np.sort


```
x = np.array([2,1,4,3,5])
np.sort(x)
```

```
x
```

如果希望用排好序的数组替代原始数组，可以使用数组的 sort 方法：

```
x.sort()
x
```

另外一个相关的函数是 argsort，该函数返回的是原始数组排好序的索
引值：

```
i = np.argsort(x)
i
```

```
x[i]
```

沿着行或列排序




```
rand = np.random.RandomState(42)
X = rand.randint(0,10,(4,6))
X
```

```
np.sort(X,axis=0)
```

```
np.sort(X,axis=1)
```
